```python
import daft
from matplotlib import rc

rc("font", family="serif", size=12)
rc("text", usetex=True)

pgm = daft.PGM()

plot_param_edge_param = {"ec": "#f89406", "linestyle": ":", "lw": 2}
plot_param_edge = {"lw": 2}
plot_param_node = {"ec": "#f89406"}
fontsize = 18

# Nodes
n = daft.Node("lambda", r"\boldmath$\lambda$\unboldmath$_k$", 1, 4.75,
              fontsize=fontsize, plot_params=plot_param_node)
n.va = "baseline"
pgm.add_node(n)
pgm.add_node("pi", r"\boldmath$\pi$", 2, 4, fontsize=fontsize, plot_params=plot_param_node)
pgm.add_node("d1", r"$\mathbf{d_1}$", 3, 5.5, fontsize=fontsize)
pgm.add_node("d2", r"$\mathbf{d_2}$", 5, 5.5, fontsize=fontsize)
pgm.add_node("d3", r"$\mathbf{d_3}$", 7, 5.5, fontsize=fontsize)
pgm.add_node("z1", r"$\mathbf{z_1}$", 3, 4, fontsize=fontsize)
pgm.add_node("z2", r"$\mathbf{z_2}$", 5, 4, fontsize=fontsize)
pgm.add_node("z3", r"$\mathbf{z_3}$", 7, 4, fontsize=fontsize)
pgm.add_node("A", r"$A$", 2, 4.55, fontsize=fontsize, plot_params=plot_param_node)
pgm.add_node("tau1", r"$\tau_1$", 2, 3, fontsize=fontsize)
pgm.add_node("tau2", r"$\tau_2$", 4, 3, fontsize=fontsize)
pgm.add_node("tau3", r"$\tau_3$", 6, 3, fontsize=fontsize)
pgm.add_node("y1", r"$y_1$", 3, 2, fontsize=fontsize, observed=True)
pgm.add_node("y2", r"$y_2$", 5, 2, fontsize=fontsize, observed=True)
pgm.add_node("y3", r"$y_3$", 7, 2, fontsize=fontsize, observed=True)
pgm.add_node("nu", r"$\nu_k$", 1, 2.2, fontsize=fontsize, plot_params=plot_param_node)
pgm.add_node("nu", r"$\nu_k$", 1, 1.8, fontsize=fontsize, plot_params=plot_param_node)
pgm.add_node("sig", r"$\sigma_k$", 1, 1.15, fontsize=fontsize, plot_params=plot_param_node)
pgm.add_node("a", r"$\mathbf{a}_k$", 1, 0.5, fontsize=fontsize, plot_params=plot_param_node)

# Auxiliary "nodes"
pgm.add_node("aux1", "", 3, 0.75, scale=0.1, fixed=True)
pgm.add_node("aux2", "", 7, 0.75, scale=0.1, fixed=True)
pgm.add_node("aux3", "", 5, 1.25, scale=0.1, fixed=True)
pgm.add_node("aux4", "", 6.2, 1.25, scale=0.1, fixed=True)

# ellipsis "nodes"
pgm.add_node("el1", "", 7.4, 5.5, scale=0.5, fixed=True)
pgm.add_node("el2", "", 7.6, 5.5, scale=0.5, fixed=True)
pgm.add_node("el3", "", 7.8, 5.5, scale=0.5, fixed=True)
pgm.add_node("el4", "", 7.4, 4, scale=0.5, fixed=True)
pgm.add_node("el5", "", 7.6, 4, scale=0.5, fixed=True)
pgm.add_node("el6", "", 7.8, 4, scale=0.5, fixed=True)
pgm.add_node("el7", "", 7.4, 3, scale=0.5, fixed=True)
pgm.add_node("el8", "", 7.6, 3, scale=0.5, fixed=True)
pgm.add_node("el9", "", 7.8, 3, scale=0.5, fixed=True)
pgm.add_node("el10", "", 7.4, 2, scale=0.5, fixed=True)
pgm.add_node("el11", "", 7.6, 2, scale=0.5, fixed=True)
pgm.add_node("el12", "", 7.8, 2, scale=0.5, fixed=True)
pgm.add_node("el13", "", 6.4, 1.25, scale=0.5, fixed=True)
pgm.add_node("el14", "", 6.6, 1.25, scale=0.5, fixed=True)
pgm.add_node("el15", "", 6.8, 1.25, scale=0.5, fixed=True)

# Edges
pgm.add_edge("lambda", "d1", plot_params=plot_param_edge_param)
pgm.add_edge("lambda", "d2", plot_params=plot_param_edge_param)
pgm.add_edge("lambda", "d3", plot_params=plot_param_edge_param)
pgm.add_edge("pi", "z1", plot_params=plot_param_edge_param)
pgm.add_edge("d1", "d2", plot_params=plot_param_edge)
pgm.add_edge("d2", "d3", plot_params=plot_param_edge)
pgm.add_edge("z1", "d1", plot_params=plot_param_edge)
pgm.add_edge("z2", "d2", plot_params=plot_param_edge)
pgm.add_edge("z3", "d3", plot_params=plot_param_edge)
pgm.add_edge("A", "z2", plot_params=plot_param_edge_param)
pgm.add_edge("A", "z3", plot_params=plot_param_edge_param)
pgm.add_edge("d1", "z2", plot_params=plot_param_edge)
pgm.add_edge("d2", "z3", plot_params=plot_param_edge)
pgm.add_edge("z1", "z2", plot_params=plot_param_edge)
pgm.add_edge("z2", "z3", plot_params=plot_param_edge)
pgm.add_edge("z1", "y1", plot_params=plot_param_edge)
pgm.add_edge("z2", "y2", plot_params=plot_param_edge)
pgm.add_edge("z3", "y3", plot_params=plot_param_edge)
pgm.add_edge("y1", "y2", plot_params=plot_param_edge)
pgm.add_edge("y2", "y3", plot_params=plot_param_edge)
pgm.add_edge("z1", "tau1", plot_params=plot_param_edge)
pgm.add_edge("z2", "tau2", plot_params=plot_param_edge)
pgm.add_edge("z3", "tau3", plot_params=plot_param_edge)
pgm.add_edge("tau1", "y1", plot_params=plot_param_edge)
pgm.add_edge("tau2", "y2", plot_params=plot_param_edge)
pgm.add_edge("tau3", "y3", plot_params=plot_param_edge)
pgm.add_edge("nu", "tau1", plot_params=plot_param_edge_param)
pgm.add_edge("nu", "tau2", plot_params=plot_param_edge_param)
pgm.add_edge("nu", "tau3", plot_params=plot_param_edge_param)
pgm.add_edge("sig", "y1", plot_params=plot_param_edge_param)
pgm.add_edge("sig", "y2", plot_params=plot_param_edge_param)
pgm.add_edge("sig", "y3", plot_params=plot_param_edge_param)
pgm.add_edge("a", "y3", plot_params=plot_param_edge_param)

# auxiliary edges
pgm.add_edge("y1", "aux1", directed=False, plot_params=plot_param_edge)
pgm.add_edge("aux1", "aux2", directed=False, plot_params=plot_param_edge)
pgm.add_edge("aux2", "y3", plot_params=plot_param_edge)
pgm.add_edge("y2", "aux3", directed=False, plot_params=plot_param_edge)
pgm.add_edge("aux3", "aux4", directed=False, plot_params=plot_param_edge)

# Plates
pgm.add_plate([0.4, 4.25, 1, 1], label=r"K", fontsize=fontsize-2)
pgm.add_plate([0.4, 0.1, 1.1, 2.15], label=r"K", fontsize=fontsize-2)

# Render
pgm.render()
```
